```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('cost_revenue_clean.csv')
```

```python
df.describe()
```

```python
X = pd.DataFrame(df, columns=["production_budget_usd"])
y = pd.DataFrame(df, columns=["worldwide_gross_usd"])
x_axis = df["production_budget_usd"].to_numpy()
y_axis = df["worldwide_gross_usd"].to_numpy()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x_axis, y_axis, alpha=0.3)
plt.title("Film Cost vs Global Revenue")
plt.xlabel("Production Budget (USD)")
plt.ylabel("Worldwide Gross (USD)")
plt.ylim(0, 1500000000)
plt.xlim(0, 350000000)
plt.show()
```

Create our regression object:

```python
regression = LinearRegression()
regression.fit(X, y)
```

-----------------------------------------------------------------------------

Slope Coefficient:

```python
regression.coef_ #theta 1
```

This slope coefficient should mean that for each dollar we spend on a movie's budget, we should expect a little more than $3 in profit. We have a positive relationship between budget and profit.

-----------------------------------------------------------------------------

Intercept:

```python
regression.intercept_ #theta 0
```

This number is a bit less realistic. It implies if we were to spend 0 dollars on a movie, then we would lose 7 million dollars.

---------------------------------------------------------------------------

```python
plt.figure(figsize=(10,6))
plt.scatter(x_axis, y_axis, alpha=0.3)
plt.plot(x_axis, regression.predict(X), color='red', linewidth=4)

plt.title("Film Cost vs Global Revenue")
plt.xlabel("Production Budget (USD)")
plt.ylabel("Worldwide Gross (USD)")
plt.ylim(0, 1500000000)
plt.xlim(0, 350000000)
plt.show()
```

```python
regression.score(X, y)
```
